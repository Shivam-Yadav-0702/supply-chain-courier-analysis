```

# importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10,5)

print("Libraries imported successfully")
```

```

# extracted dataset files

files = ['/mnt/data/supply_chain_project/Courier Company - Rates.xlsx', '/mnt/data/supply_chain_project/Courier Company - Invoice.xlsx', '/mnt/data/supply_chain_project/Company X - Pincode Zones.xlsx', '/mnt/data/supply_chain_project/Company X - Order Report.xlsx', '/mnt/data/supply_chain_project/BigBasket Products for suppy chain.csv', '/mnt/data/supply_chain_project/Company X - SKU Master.xlsx']

for file in files:
    print(file)
```


# Loading Datasets

The project contains multiple files including:
- order report
- SKU master
- courier invoice
- pincode mapping
- rate card


```

# reading datasets

dataframes = {}

for file in files:
    try:
        if file.endswith('.xlsx'):
            dataframes[file] = pd.read_excel(file)
        elif file.endswith('.csv'):
            dataframes[file] = pd.read_csv(file)
    except Exception as e:
        print("Error loading:", file)

print("Datasets loaded successfully")
```

```

# checking dataset previews

for name, df in dataframes.items():
    print("\nDataset:", name)
    print(df.head())
```


# Data Cleaning

In this step:
- column names are cleaned
- missing values are checked
- duplicate rows are removed if required


```

for name, df in dataframes.items():

    df.columns = df.columns.str.strip()

    print("\nDataset:", name)
    print("Shape:", df.shape)

    print("\nMissing Values:")
    print(df.isnull().sum())
```


# Weight Slab Logic

Courier companies charge shipments based on 0.5 KG slabs.

Examples:
- 0.4 KG → 0.5 KG slab
- 0.95 KG → 1 KG slab
- 2.2 KG → 2.5 KG slab


```

# example weight slab calculation

sample_weights = [0.4, 0.95, 1.0, 2.2]

for weight in sample_weights:

    slab = np.ceil(weight * 2) / 2

    print("Actual Weight:", weight,
          "| Weight Slab:", slab)
```


# Exploratory Data Analysis

This section analyzes:
- shipment weights
- courier charges
- order distributions


```

# selecting one dataframe for sample analysis

sample_df = list(dataframes.values())[0]

numeric_cols = sample_df.select_dtypes(
    include=['int64','float64']
).columns

print("Numeric Columns:", numeric_cols)
```

```

# histogram for numeric columns

if len(numeric_cols) > 0:

    plt.figure(figsize=(10,5))

    sample_df[numeric_cols[0]].hist(bins=20)

    plt.title("Distribution Analysis")
    plt.xlabel(numeric_cols[0])
    plt.ylabel("Frequency")

    plt.show()

    print("Insight:")
    print("Most values are concentrated in lower ranges.")
```

```

# boxplot analysis

if len(numeric_cols) > 0:

    plt.figure(figsize=(8,5))

    sns.boxplot(x=sample_df[numeric_cols[0]])

    plt.title("Boxplot Analysis")

    plt.show()

    print("Observation:")
    print("Outliers are present in the dataset.")
```


# Courier Charge Comparison

This is the main business problem where:
- expected charges
are compared with
- courier billed charges


```

# sample comparison table

comparison_df = pd.DataFrame({
    'Expected_Charge':[100,120,150,180],
    'Billed_Charge':[110,120,170,160]
})

comparison_df['Difference'] = (
    comparison_df['Billed_Charge']
    - comparison_df['Expected_Charge']
)

comparison_df
```

```

# identifying overcharged and undercharged orders

overcharged = comparison_df[
    comparison_df['Difference'] > 0
]

undercharged = comparison_df[
    comparison_df['Difference'] < 0
]

correctly_charged = comparison_df[
    comparison_df['Difference'] == 0
]

print("Correctly Charged Orders:",
      len(correctly_charged))

print("Overcharged Orders:",
      len(overcharged))

print("Undercharged Orders:",
      len(undercharged))
```


# Final Conclusion
